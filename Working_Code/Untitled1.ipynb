```python
import pandas as pd
```

```python
df = pd.read_csv('/data/g0017139/CNA/enrichment_matrix_Gene Ontology - Biolocal Processes.tsv',sep='\t')
df[['gene_set_name', 'Link']] = df['Unnamed: 0'].str.split(' --',expand=True)
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df1 = pd.read_csv('/data/g0017139/CNA/gobp_immune_gene_sets_manualselection_updated.tsv', sep='\t')
df1 = df1[df1['is_immune_process'] == True]
```

```python
df = pd.merge(df, df1, left_on='gene_set_name', right_on='gene_set_name',how='left')
```

From GO-BP, all gene set were selected that contained the motif â€˜METABOLIC_PROCESSâ€™ in the title. In addition, gene sets that contained the name of a metabolite or class of metabolites in combination with the motif â€˜_TRANSPORTâ€™ in the title were selected. Furthermore, gene sets not containing these title motifs, but nevertheless associated with (cancer) metabolism, were manually selected on the basis of metabolic pathway names

```python
names = pd.read_csv('/data/g0017139/CNA/metabolic_gene_sets.txt', sep='\t')['GO_NUCLEOBASE_METABOLIC_PROCESS']
df['is_methabolic_process'] = df['gene_set_name'].isin(names.str.replace('GO','GOBP'))
```

```python
df.head()
```

```python
print(f'is_methabolic_process = {df["is_methabolic_process"].sum()}')
print(f'is_immune_process = {df["is_immune_process"].sum()}')
print(f'total = {len(df)}')
```
