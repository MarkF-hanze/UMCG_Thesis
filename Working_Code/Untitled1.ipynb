```python
import pandas as pd
```

# GO BP

```python
df = pd.read_csv('/data/g0017139/CNA/enrichment_matrix_Gene Ontology - Biolocal Processes.tsv',sep='\t')
df[['gene_set_name', 'Link']] = df['Unnamed: 0'].str.split(' --',expand=True)
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
# Immune
df1 = pd.read_csv('/data/g0017139/CNA/gobp_immune_gene_sets_manualselection_updated.tsv', sep='\t')
df1 = df1[df1['is_immune_process'] == True]
df = pd.merge(df, df1, left_on='gene_set_name', right_on='gene_set_name',how='left')
```

From GO-BP, all gene set were selected that contained the motif â€˜METABOLIC_PROCESSâ€™ in the title. In addition, gene sets that contained the name of a metabolite or class of metabolites in combination with the motif â€˜_TRANSPORTâ€™ in the title were selected. Furthermore, gene sets not containing these title motifs, but nevertheless associated with (cancer) metabolism, were manually selected on the basis of metabolic pathway names

```python
# Metabolic
names = pd.read_csv('/data/g0017139/CNA/metabolic_gene_sets.txt', sep='\t')['GO_NUCLEOBASE_METABOLIC_PROCESS']
df['is_methabolic_process'] = df['gene_set_name'].isin(names.str.replace('GO','GOBP'))
```

```python
#DNA Repair
immune = pd.read_excel('/data/g0017139/CNA/GOBP_REACTOME_RELEVANT_GENESETS.xlsx',header=None)
names_immune = immune[0]

df['is_DNA_Repair'] = df['gene_set_name'].isin(names_immune)
```

```python
print(f'is_methabolic_process = {df["is_methabolic_process"].sum()}')
print(f'is_immune_process = {df["is_immune_process"].sum()}')
print(f'is_DNA_Repair = {df["is_DNA_Repair"].sum()}')
print(f'total = {len(df)}')
```

```python
df['keep'] = ((df['is_immune_process']==True) | (df['is_methabolic_process']==True) | (df['is_DNA_Repair']==True))
analysis_df = df[df['keep']==True]
```

```python
df.head()
```

```python
import scipy.stats as st
```

```python
test_df = df.drop(['gene_set_name','Link','is_immune_process', 'is_methabolic_process','is_DNA_Repair',  'keep'],
                 axis=1)
shape = test_df.shape
value = st.norm.ppf(0.05/(shape[0]*shape[1]))
test_df = test_df.abs()
test_df = (test_df > abs(value))
df['Any_Sig'] = test_df.sum(axis=1) > 0
test_df = test_df[test_df.sum(axis=1) > 0]
significant_conponents = {}
for index, row in test_df.iterrows():
    significant_conponents[index] = list(row[row==True].index)
```

```python
significant_conponents
```

```python
biology_df = df[((df['Any_Sig'] == True) & (df['keep'] == True))]
```

```python
biology_df.loc[:,"Significant"] = pd.Series(significant_conponents)
biology_df = biology_df.fillna(False)
```

```python
counts = {}
imp_rows = ['is_immune_process', 'is_methabolic_process', 'is_DNA_Repair']
for index, row in biology_df.iterrows():
    for comp in row['Significant']:
        if comp not in counts:
            counts[comp] = {
                'is_immune_process': 0,
                'is_methabolic_process': 0,
                'is_DNA_Repair': 0
            }
        for imp_row in imp_rows:
            counts[comp][imp_row] += row[imp_row]
```

```python
counts
```
