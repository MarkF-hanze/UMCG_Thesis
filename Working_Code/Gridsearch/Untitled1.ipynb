```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
df = pd.read_csv('/data/g0017139/Set2/TCGA__RSEM_genes_RNAseq__duplicate_samples_removed__genes_with_all_zeroes_removed.txt',
                               sep=None, engine='python', header=0,
)
df = df.T
df.columns = [str(x) for x in df.columns]
df.to_parquet("/data/g0017139/TCGA__RSEM.parquet")

scaler = StandardScaler()
df_norm = scaler.fit_transform(df)
df_norm = pd.DataFrame(df_norm)
df_norm.columns = [str(x) for x in df_norm.columns]
df_norm.to_csv("/data/g0017139/TCGA__RSEM_norm.dat", sep = " ",header=False, index=False)
df_norm.to_parquet('/data/g0017139/TCGA__RSEM_norm.parquet')
```
