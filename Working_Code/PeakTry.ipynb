```python
#Load DNA Repair
dna_df = pd.read_excel('/data/g0017139/CNA/GOBP_REACTOME_RELEVANT_GENESETS.xlsx',header=None)
```

```python
# Merge evertying with the results of df
df = pd.merge(df, immune_df, left_on='gene_set_name', right_on='gene_set_name',how='left')
df['is_methabolic_process'] = df['gene_set_name'].isin(metabolic)
df['is_DNA_Repair'] = df['gene_set_name'].isin(dna_df[0])

# Only keep that has at least one True
df['keep'] = ((df['is_immune_process']==True) | (df['is_methabolic_process']==True) | (df['is_DNA_Repair']==True))
analysis_df = df[df['keep']==True]
```

```python
print(f'is_methabolic_process = {df["is_methabolic_process"].sum()}')
print(f'is_immune_process = {df["is_immune_process"].sum()}')
print(f'is_DNA_Repair = {df["is_DNA_Repair"].sum()}')
print(f'To keep = {df["keep"].sum()}')
print(f'total = {len(df)}')
```

```python
import scipy.stats as st
```

```python
# Drop the not relevant genes
test_df = df.drop(['gene_set_name','Link','is_immune_process', 'is_methabolic_process','is_DNA_Repair',  'keep'],
                 axis=1)
shape = test_df.shape
# What is the number of tests? 
# Bonferroni correction
value = st.norm.ppf(0.05/(shape[0]*shape[1]))
test_df = test_df.abs()
test_df = (test_df > abs(value))

# See of any value in the row is still significant (So higher than corrected value) and only keep those
df['Any_Sig'] = test_df.sum(axis=1) > 0
test_df = test_df[test_df.sum(axis=1) > 0]
# Loop over the significant rows and get those column names (The names of the components)
significant_conponents = {}
for index, row in test_df.iterrows():
    significant_conponents[index] = list(row[row==True].index)

# Only keep the rows that have both a signifant value and this signicant value is in a relevent gene for the gene sets
biology_df = df[((df['Any_Sig'] == True) & (df['keep'] == True))]
# Add the signifcant components and change Nan to False
biology_df.loc[:,"Significant"] = pd.Series(significant_conponents)
biology_df = biology_df.fillna(False)

# Count how often each indipendent components has a gene that is linked to a geneset that is linked to a process
counts = {}
imp_rows = ['is_immune_process', 'is_methabolic_process', 'is_DNA_Repair']
for index, row in biology_df.iterrows():
    for comp in row['Significant']:
        if comp not in counts:
            counts[comp] = {
                'is_immune_process': 0,
                'is_methabolic_process': 0,
                'is_DNA_Repair': 0
            }
        for imp_row in imp_rows:
            counts[comp][imp_row] += row[imp_row]
```
